from describe above we've concluded that the only possible way to tackle the missing values from the three columns(Price Per Unit	Quantity	Total Spent) is mean for price per unit and quantity. For Total spent we will use mode so that it can well handle the roboust outliers.


```python
# Convert potentially-non-numeric columns to numeric (coerce errors to NaN), then fill missing values
# This avoids FutureWarnings about numeric-only aggregation and ensures correct calculations
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")

# Fill missing with mean/mode after conversion
df["Price Per Unit"].fillna(df["Price Per Unit"].mean(), inplace=True)
df["Quantity"].fillna(df["Quantity"].mean(), inplace=True)
# Use mode if available, otherwise fill with 0 as fallback
mode_vals = df["Total Spent"].mode()
df["Total Spent"].fillna(mode_vals.iloc[0] if not mode_vals.empty else 0, inplace=True)

# Check remaining nulls
df.isnull().sum()
```
